```python
from PIL import Image
```

```python
# Open an image file
# ps. you can also specify an URL
pil_image = Image.open("stark.jpeg")
# image = Image.open(r"path.jpg") # specify r for Windows

# Image information
print("Image format :", pil_image.format)
print("Image mode :", pil_image.mode)
print("Image size :", pil_image.size)
```

```python
# Convert into grayscale
gray_image = pil_image.convert("L")
gray_image
```
