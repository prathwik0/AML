```python
import numpy as np
import pandas as pd
```

```python
data = {
    "animal": [
        "cat",
        "cat",
        "snake",
        "dog",
        "dog",
        "cat",
        "snake",
        "cat",
        "dog",
        "dog",
    ],
    "age": [2.5, 3, 0.5, np.nan, 5, 2, 4, np.nan, 7, 4],
    "visits": [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    "priority": ["yes", "yes", "no", "yes", "no", "no", "no", "yes", "no", "no"],
}

df = pd.DataFrame(data)
df
```

----
## Drop na method
Drop method works well for very large datasets with few missing values. <br>
But if you are looking at smaller datasets such as iris, you should probably consider other options.

```python
df.dropna()
```

----
## Fill na method

```python
mean = df.loc[:, "age"].mean()
df.fillna(mean)
```
